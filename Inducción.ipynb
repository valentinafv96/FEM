```python
#Importando librerías

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
```

```python
#Definiendo constantes

B = 1.253e-3         #Campo magnético en [T]
ω = 4.5              #Velocidad angular en [1/s]

#Definiendo variables 

t = np.linspace(0,4,200)                 #Tiempo en [s]

r = np.array([2.1,1.4,2.1])*1e-2         #Radio del embobinado [m]
N = np.array([64,107,175])               #Número de espiras
```

```python
datos = {'N. Espiras (N)': N,
         'Radio (r)': r}

df=pd.DataFrame(datos, columns=['N. Espiras (N)','Radio (r)'])

df
```

```python
#Cálculo del flujo magnpetico a través de la espira y la inducción correspondiente

def MFlux (r): return B*np.pi*pow(r,2)*np.cos(ω*t)  

def induction (N,r,t): return N*B*np.pi*pow(r,2)*ω*np.sin(ω*t)

#Corriente inducida máxima

FEMax = induction(N,r,np.pi*0.5*(1/ω)) 
```

```python
#Graficando datos

plt.figure(figsize=(20,12))

plt.subplot(211)

plt.title('\nFlujo magnético\n', fontsize = 18, fontstyle= 'italic', fontweight= 'bold')

plt.plot(t,MFlux(r[0]), 'b-', label='N = %.2f, r = %.2f'%(N[0],r[0]))
plt.plot(t,MFlux(r[1]), 'r-', label='N = %.2f, r = %.2f'%(N[1],r[1]))
plt.plot(t,MFlux(r[2]), 'g-', label='N = %.2f, r = %.2f'%(N[2],r[2]))

plt.xlabel('Tiempo [s]\n', fontsize=16, fontstyle= 'italic')
plt.ylabel('$\Phi_{B} \quad [Wb]$', fontsize=16, fontstyle= 'italic')

plt.ticklabel_format(axis='y', style='sci', scilimits=(0,0))

plt.xlim(0,max(t))
plt.ylim(-4e-6,4e-6)

plt.grid(alpha=0.5)

plt.legend(fontsize=16)

#--------------------------------------------------------------------------------------

plt.subplot(212)

plt.title('\n\nInducción magnética\n', fontsize = 18, fontstyle= 'italic', fontweight= 'bold')

plt.plot(t,induction(N[0],r[0],t),'b--', label='N = %.2f, r = %.2f'%(N[0],r[0]))
plt.plot(t,induction(N[1],r[1],t),'r--', label='N = %.2f, r = %.2f'%(N[1],r[1]))
plt.plot(t,induction(N[2],r[2],t),'g--', label='N = %.2f, r = %.2f'%(N[2],r[2]))
plt.plot(t,FEMax[2]*np.ones(len(t)),'k', label='$\epsilon_{MAX} = %E$'%(FEMax[2]))

plt.xlabel('Tiempo [s]', fontsize=16, fontstyle= 'italic')
plt.ylabel('$\epsilon \quad [V]$ ', fontsize=16, fontstyle= 'italic')

plt.ticklabel_format(axis='y', style='sci', scilimits=(0,0))

plt.xlim(0,max(t))
plt.ylim(-4e-3,4e-3)

plt.grid(alpha=0.5)

plt.legend(fontsize=14)

plt.subplots_adjust(hspace = 0.4)

plt.show()
```
